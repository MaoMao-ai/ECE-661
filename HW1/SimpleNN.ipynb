### Code block 2: Useful classes
Customized implementation of [nn.Conv2d](https://pytorch.org/docs/stable/generated/torch.nn.Conv2d.html) and [nn.Linear](https://pytorch.org/docs/stable/generated/torch.nn.Linear.html#torch.nn.Linear), do not change.

```python
class CONV(nn.Conv2d):
    def __init__(self, in_channels, out_channels, kernel_size, stride=1,
                 padding=0, dilation=1, groups=1,
                 bias=False, padding_mode='zeros'):
        self.input = None
        self.output = None
        kernel_size = _pair(kernel_size)
        stride = _pair(stride)
        padding = _pair(padding)
        dilation = _pair(dilation)
        super(CONV, self).__init__(
            in_channels, out_channels, kernel_size, stride, padding, dilation, 
            groups, bias, padding_mode)
    
    def forward(self, input):
        self.input = input
        # 'pytorch 2.0'
        self.output = super().forward(input)
        return self.output
    
class FC(nn.Linear):
    def __init__(self, in_features, out_features, bias=True):
        self.input = None
        self.output = None
        super(FC, self).__init__(in_features, out_features, bias)
    
    def forward(self, input):
        self.input = input
        self.output = F.linear(input, self.weight, self.bias)
        return self.output
```

### Code block 3: SimpleNN implementation

```python
""" 
Lab 2(a)
Build the SimpleNN model by following Table 1
"""

# Create the neural network module: LeNet-5
class SimpleNN(nn.Module):
    def __init__(self):
        super(SimpleNN, self).__init__()
        # Layer definition
        self.conv1 = CONV(in_channels=3, out_channels=16, kernel_size=5, padding=2)    #Your code here
        self.pool1 = nn.MaxPool2d(kernel_size=4, stride=2) 
        self.conv2 = CONV(in_channels=16, out_channels=16, kernel_size=3, padding=2)   #Your code here
        self.pool2 = nn.MaxPool2d(kernel_size=3, stride=2)
        self.conv3 = CONV(in_channels=16, out_channels=32, kernel_size=7, padding=2)   #Your code here
        self.pool3 = nn.MaxPool2d(kernel_size=2, stride=2)
        self.fc1   = FC(in_features=288, out_features=32)    #Your code here
        self.fc2   = FC(in_features=32, out_features=10)    #Your code here

    def forward(self, x):
        # Forward pass computation
        # Conv 1
           x = F.relu(self.conv1(x))  #Your code here
        # MaxPool
           x = self.pool1(x) #Your code here
        # Conv 2
           x = F.relu(self.conv2(x)) #Your code here
        # MaxPool
           x = self.pool2(x) #Your code here
        # Conv 3
           x = F.relu(self.conv3(x)) #Your code here
        # MaxPool
           x = self.pool3(x) #Your code here
        # Flatten
           x = torch.flatten(x, 1) #Your code here
        # FC 1
           x = F.relu(self.fc1(x)) #Your code here
        # FC 2
           out = F.relu(self.fc2(x)) #Your code here
           return out

# GPU check                
device = 'cuda' if torch.cuda.is_available() else 'cpu'
if device =='cuda':
    print("Run on GPU...")
else:
    print("Run on CPU...")

# Model Definition  
net = SimpleNN()
net = net.to(device)

# Test forward pass
data = torch.randn(5,3,32,32)
data = data.to(device)
# Forward pass "data" through "net" to get output "out" 
out = net(data)   #Your code here

# Check output shape
assert(out.detach().cpu().numpy().shape == (5,10))
print("Forward pass successful")
```

```python
"""
Lab 2(b)
"""
# Forward pass of a single image
data = torch.randn(1,3,32,32).to(device)
# Forward pass "data" through "net" to get output "out" 
out = net(data)    #Your code here

# Iterate through all the CONV and FC layers of the model
for name, module in net.named_modules():
    if isinstance(module, CONV) or isinstance(module, FC):
        # Get the input feature map of the module as a NumPy array
        input = module.input.cpu().detach().numpy()      #Your code here
        # Get the output feature map of the module as a NumPy array
        output = module.output.cpu().detach().numpy()     #Your code here
        # Get the weight of the module as a NumPy array
        weight = module.weight.cpu().detach().numpy()     #Your code here
        # Compute the number of parameters in the weight
        num_Param = weight.size     #Your code here
        # Compute the number of MACs in the layer
        # For convolutional layers, MACs = (input channels * kernel height * kernel width * output height * output width * output channels)
        # For fully connected layers, MACs = (input units * output units)
        if isinstance(module, CONV):
            # Get the dimensions of the input and output feature maps
            in_channels, in_h, in_w = input.shape[1:]
            out_channels, out_h, out_w = output.shape[1:]
            kernel_h, kernel_w = module.weight.shape[2:]  # Kernel size
            
            # MACs for a convolutional layer
            num_MAC = in_channels * kernel_h * kernel_w * out_h * out_w * out_channels
        
        elif isinstance(module, FC):
            # MACs for a fully connected layer
            in_units = input.shape[1]
            out_units = output.shape[1]
            num_MAC = in_units * out_units
        
        print(f'{name:10} {str(input.shape):20} {str(output.shape):20} {str(weight.shape):20} {str(num_Param):10} {str(num_MAC):10}')
```
